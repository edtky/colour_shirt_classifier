# Custom Image Training of shirt color classifier

---------

1. Split data into approximately 80% train, 20% test. Each class has about 300 images.

2. Store train and test images in a folder according to this structure:

    - shirts//  
        - train//  
            - red//  
            - blue//  
            - green//  
        - test//  
            - red//  
            - blue//  
            - green//  
            
            
3. Modify input of `model_trainer.trainModel()`:
    - num_objects: number of classes in your dataset
    - num_experiments: epoch, number of complete passes through your training set
    
    
4. About **50 epoch** was required to consistently hit about 90% accuracy on train and test.
  
  
  
Acknowledgement:
https://github.com/OlafenwaMoses/ImageAI/blob/master/imageai/Prediction/CUSTOMTRAINING.md


```python
from imageai.Prediction.Custom import ModelTraining
model_trainer = ModelTraining()
model_trainer.setModelTypeAsResNet()
model_trainer.setDataDirectory("data/shirts")
model_trainer.trainModel(num_objects=3, num_experiments=100, enhance_data=True, batch_size=32, show_network_summary=True)

```
