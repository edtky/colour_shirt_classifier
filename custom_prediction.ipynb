```python
from imageai.Prediction.Custom import CustomImagePrediction
from IPython.display import Image
import os
execution_path = os.getcwd()
```

```python
# declare test images
image1 = 'data/test/test7.jpg'
image2 = 'data/test/test2.jpg'
image3 = 'data/test/test3.jpg'
image4 = 'data/test/test4.jpg'
image5 = 'data/test/test6.jpg'
image6 = 'data/test/test9.jpg'

good_examples = [image7, image2, image3]
bad_examples = [image4, image5, image6]
```

```python
# load model
best_model = "data/shirts/models/model_ex-062_acc-0.987500.h5"
model_class = "data/shirts/json/model_class.json"

prediction = CustomImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath(os.path.join(execution_path, best_model))
prediction.setJsonPath(os.path.join(execution_path, model_class))
prediction.loadModel(num_objects=3)
```

```python
print("GOOD EXAMPLES:")

for test_image in good_examples:
    
    display(Image(filename=test_image, width=300))

    predictions, probabilities = prediction.predictImage(os.path.join(execution_path, test_image), result_count=5)

    for eachPrediction, eachProbability in zip(predictions, probabilities):
        print(eachPrediction + " : " + eachProbability)
    
    print("\n")
    
```

```python
print("BAD EXAMPLES")

for test_image in bad_examples:
    
    display(Image(filename=test_image, width=300))

    predictions, probabilities = prediction.predictImage(os.path.join(execution_path, test_image), result_count=5)

    for eachPrediction, eachProbability in zip(predictions, probabilities):
        print(eachPrediction + " : " + eachProbability)
    
    print("\n")
```
